```python
import numpy as np
```

```python
f1 = open("build/nlm_gpu1.txt", "r")
nlm1 = f1.readlines()
ln = int(len(nlm1)/4)
nlm = np.array([np.array([nlm1[i]]).astype(np.double) for i in range(len(nlm1))])
nlm = np.array([nlm[:ln], nlm[ln:2*ln],  nlm[2*ln:3*ln],  nlm[3 * ln:4*ln]])
nlm = nlm.reshape(4, ln)
f1.close()
```

```python
import matplotlib.pyplot as plt
colors =['ro-','bo-','go-','co-','yo-','ko-','wo-','mo-']
ds = [2,3,4,5]
labels = ['conv_size = ' + str(ds[i]) for i in range(len(ds))]
#plt.yscale('log')
#plt.xscale('log')

plt.xlabel('Radius parameter')
plt.ylabel('Time in s')
#plt.ylim([1e-4, 1e4])
plt.xlim([1, 8])
radius =  [2,3,4,5,6]
for i in range(len(ds)):
    plt.plot(radius, nlm[i,:], colors[i% len(colors)])
plt.legend(labels)
plt.title("Time measurement in non local means for Lena 256 * 256 ")
plt.show()    
```

```python
f1 = open("build/nlm_gpu2.txt", "r")
nlm1 = f1.readlines()
ln = int(len(nlm1)/4)
nlm = np.array([np.array([nlm1[i]]).astype(np.double) for i in range(len(nlm1))])
nlm = np.array([nlm[:ln], nlm[ln:2*ln],  nlm[2*ln:3*ln],  nlm[3 * ln:4*ln]])
nlm = nlm.reshape(4, ln)
f1.close()
```

```python
import matplotlib.pyplot as plt
colors =['ro-','bo-','go-','co-','yo-','ko-','wo-','mo-']
ds = [2,3,4,5]
labels = ['conv_size = ' + str(ds[i]) for i in range(len(ds))]


plt.xlabel('Radius parameter')
plt.ylabel('Time in s')

plt.xlim([1, 8])
radius =  [2,3,4,5,6]
for i in range(len(ds)):
    plt.plot(radius, nlm[i,:], colors[i% len(colors)])
plt.legend(labels)
plt.title("Time measurement in non local means for Lena 514 * 514 ")
plt.show()    
```

```python
f1 = open("build/knn_gpu1.txt", "r")
f2 = open("build/knn_gpu2.txt", "r")
nlm1 = f1.readlines() + f2.readlines()
ln = int(len(nlm1)/2)
nlm = np.array([np.array([nlm1[i]]).astype(np.double) for i in range(len(nlm1))])
nlm = nlm.reshape(2, ln)
knn = nlm
f1.close()
```

```python
import matplotlib.pyplot as plt
colors =['ro-','bo-','go-','co-','yo-','ko-','wo-','mo-']
ds = [256, 512]
labels = ['Image_size = ' + str(ds[i]) +' * '+str(ds[i]) for i in range(len(ds))]
#plt.yscale('log')
#plt.xscale('log')

plt.xlabel('number of neighbours')
plt.ylabel('Time in s')
#plt.ylim([1e-4, 1e4])
plt.xlim([1, 18])
neigh =  [2 + 2 * i for i in range(7) ]
for i in range(len(ds)):
    plt.plot(neigh, knn[i,:], colors[i% len(colors)])
plt.legend(labels)
plt.title("Time measurement for knn GPU")
plt.show()    
```

```python
f1 = open("build/knn_shared_gpu1.txt", "r")
f2 = open("build/knn_shared_gpu2.txt", "r")
nlm1 = f1.readlines() + f2.readlines()
ln = int(len(nlm1)/2)
nlm = np.array([np.array([nlm1[i]]).astype(np.double) for i in range(len(nlm1))])
nlm = nlm.reshape(2, ln)
knn = nlm
f1.close()
```

```python
import matplotlib.pyplot as plt
colors =['ro-','bo-','go-','co-','yo-','ko-','wo-','mo-']
ds = [256, 512]
labels = ['Image_size = ' + str(ds[i]) +' * '+str(ds[i]) for i in range(len(ds))]
#plt.yscale('log')
#plt.xscale('log')

plt.xlabel('number of neighbours')
plt.ylabel('Time in s')
#plt.ylim([1e-4, 1e4])
plt.xlim([1, 18])
neigh =  [2 + 2 * i for i in range(5) ]
for i in range(len(ds)):
    plt.plot(neigh, knn[i,:], colors[i% len(colors)])
plt.legend(labels)
plt.title("Time measurement for knn GPU Shared")
plt.show()    
```
